```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("train.csv")
```

```python
data.head()
```

```python
data['Sex']=[1 if i=='male' else 0 for i in data['Sex']]
```

```python
data.drop(data[data['Embarked'].isnull()].index, inplace=True)
```

```python
data['Age'].fillna(value=data['Age'].mean(), inplace=True)
```

```python
x_train=data[['Pclass', 'Age', 'Sex', 'SibSp', 'Parch']]
```

```python
y_train=data['Survived']
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
classifier=LogisticRegression()
```

```python
classifier.fit(x_train, y_train)
```

```python
data2=pd.read_csv("test.csv")
```

```python
data2.head()
```

```python
data2['Age'].fillna(value=data2['Age'].mean(), inplace=True)
```

```python
data2['Sex']=[1 if i=='male' else 0 for i in data2['Sex']]
```

```python
x_test=data2[['Pclass', 'Age', 'Sex', 'SibSp', 'Parch']]
```

```python
y_res=classifier.predict(x_test)
```

```python
y_ans=pd.read_csv('gender_submission.csv')['Survived']
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_ans, y_res)
```
